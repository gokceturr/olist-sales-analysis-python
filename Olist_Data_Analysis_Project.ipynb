 # Kargo Suresi Istatistikleri


```python
# Convert timestamp columns to datetime objects
# timestamp to datetime
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'])
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'])

# Calculate delivery time (actual delivery date - order purchase date)
orders_df['delivery_time'] = (orders_df['order_delivered_customer_date'] - orders_df['order_purchase_timestamp']).dt.days

# Calculate shipping time (actual delivery date - estimated delivery date) (MÃ¼ÅŸteriye gerÃ§ek sipariÅŸ teslim tarihi - SatÄ±n alma anÄ±nda mÃ¼ÅŸteriye bildirilen tahmini teslimat tarihi)
orders_df['shipping_time'] = (orders_df['order_delivered_customer_date'] - orders_df['order_estimated_delivery_date']).dt.days

# Calculate descriptive statistics for delivery time
delivery_time_mean = orders_df['delivery_time'].mean()
delivery_time_median = orders_df['delivery_time'].median() # Ortadaki teslimat sÃ¼resini temsil eder.
delivery_time_std = orders_df['delivery_time'].std() # Teslimat sÃ¼relerindeki deÄŸiÅŸkenliÄŸi Ã¶lÃ§er.

print(f"Ortalama Teslimat SÃ¼resi: {delivery_time_mean:.2f} gÃ¼n")
print(f"Medyan Teslimat SÃ¼resi: {delivery_time_median:.2f} gÃ¼n")
print(f"Teslimat SÃ¼resi Standart SapmasÄ±: {delivery_time_std:.2f} gÃ¼n")

# Calculate descriptive statistics for shipping time
shipping_time_mean = orders_df['shipping_time'].mean()
shipping_time_median = orders_df['shipping_time'].median()
shipping_time_std = orders_df['shipping_time'].std()

print(f"\nOrtalama Kargo SÃ¼resi: {shipping_time_mean:.2f} gÃ¼n")
print(f"Medyan Kargo SÃ¼resi: {shipping_time_median:.2f} gÃ¼n")
print(f"Kargo SÃ¼resi Standart SapmasÄ±: {shipping_time_std:.2f} gÃ¼n")
```

MÃ¼ÅŸteri memnuniyeti puanlarÄ±nÄ±n ortalama, medyan deÄŸerlerini hesapladik. Memnuniyet puanlarÄ±nÄ± (4 ve Ã¼zeri ile 4 altÄ±) karÅŸÄ±laÅŸtÄ±rdik ve genel memnuniyet oranÄ±nÄ± bulduk.

 # Korelasyonlar

## Veri birleÅŸtirme

Korelasyon analizi iÃ§in gerekli olan `orders_df` ve `reviews_df` DataFrame'lerini birleÅŸtirdik. AyrÄ±ca, sonrasinda `order_payments` ve `order_items` veri setlerini merge ettik. .


```python
# Merge orders and reviews dataframes
merged_df = pd.merge(orders_df, reviews_df, on='order_id', how='inner')

# Load payments and items dataframes
payments_df = pd.read_csv('/content/olist_order_payments_dataset.csv')
items_df = pd.read_csv('/content/olist_order_items_dataset.csv')

# Merge payments and items dataframes with the merged dataframe
merged_df = pd.merge(merged_df, payments_df, on='order_id', how='left')
merged_df = pd.merge(merged_df, items_df, on='order_id', how='left')

# Display the first few rows of the final merged dataframe
display(merged_df.head())
```

## Veri hazÄ±rlÄ±ÄŸÄ±
Korelasyon analizi iÃ§in ilgili sÃ¼tunlarÄ± seÃ§tik ve gereken yerlerde veri tÃ¼rlerini dÃ¶nÃ¼ÅŸtÃ¼rduk. Eksik deÄŸerleri inceledik.


```python
# Select columns for correlation analysis
correlation_cols = ['delivery_time', 'shipping_time', 'review_score', 'price', 'freight_value']
correlation_df = merged_df[correlation_cols].copy()

# Handle missing values by dropping rows with NaNs in the selected columns
correlation_df.dropna(inplace=True)

# Display the first few rows and info of the correlation dataframe
print("Correlation DataFrame - Ä°lk 5 SatÄ±r:")
display(correlation_df.head())

print("\nCorrelation DataFrame - Bilgi:")
correlation_df.info()
```

## Korelasyon hesaplama
SeÃ§ilen deÄŸiÅŸkenler arasÄ±ndaki korelasyon matrisini hesapladik.


```python
correlation_matrix = correlation_df.corr()
print("Korelasyon Matrisi:")
# Display the correlation matrix with 3 decimal places
display(correlation_matrix.style.format("{:.3f}"))
```

## Korelasyon gÃ¶rselleÅŸtirme
Korelasyon matrisini bir Ä±sÄ± haritasÄ± (`heatmap`) ile gÃ¶rselleÅŸtirerek iliÅŸkileri daha anlaÅŸÄ±lÄ±r hale getirdik.


```python
# ðŸ€
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('DeÄŸiÅŸkenler ArasÄ± Korelasyon IsÄ± HaritasÄ±')
plt.show()
```

*   delivery_time ile shipping_time arasÄ±ndaki deÄŸer 0.592 pozitif ve orta dÃ¼zeyde bir iliÅŸki olduÄŸunu gÃ¶sterir. Teslimat sÃ¼resi arttÄ±kÃ§a kargo sÃ¼resi de artma eÄŸilimindedir (ki bu mantÄ±klÄ±dÄ±r, Ã§Ã¼nkÃ¼ kargo sÃ¼resi teslimat sÃ¼resinin bir parÃ§asÄ±dÄ±r).
*   shipping_time ile review_score arasÄ±ndaki deÄŸer -0.228 zayÄ±f negatif bir iliÅŸki olduÄŸunu gÃ¶sterir. Kargo sÃ¼resi uzadÄ±kÃ§a mÃ¼ÅŸteri memnuniyet puanÄ±nÄ±n bir miktar dÃ¼ÅŸme eÄŸiliminde olduÄŸunu dÃ¼ÅŸÃ¼ndÃ¼rÃ¼r.
*   price ile review_score arasÄ±ndaki deÄŸer 0.004 Ã§ok zayÄ±f, neredeyse sÄ±fÄ±r bir iliÅŸki olduÄŸunu gÃ¶sterir. Bu veri setine gÃ¶re, Ã¼rÃ¼n fiyatÄ±nÄ±n mÃ¼ÅŸteri memnuniyet puanÄ± Ã¼zerinde belirgin bir doÄŸrusal etkisi yoktur.


## Ä°liÅŸkileri keÅŸfetme ve aÃ§Ä±klama

```python
# 1. Interpret correlation between 'price' and 'delivery_time' - siparis fiyati ve teslim suresi
price_delivery_corr = correlation_matrix.loc['price', 'delivery_time']
print(f"Korelasyon (price vs. delivery_time): {price_delivery_corr:.2f}")
if abs(price_delivery_corr) < 0.1:
    print("Yorum: Fiyat ve teslimat sÃ¼resi arasÄ±nda Ã§ok zayÄ±f bir iliÅŸki vardÄ±r.")
elif abs(price_delivery_corr) < 0.4:
    print("Yorum: Fiyat ve teslimat sÃ¼resi arasÄ±nda zayÄ±f bir iliÅŸki vardÄ±r.")
elif abs(price_delivery_corr) < 0.7:
    print("Yorum: Fiyat ve teslimat sÃ¼resi arasÄ±nda orta dÃ¼zeyde bir iliÅŸki vardÄ±r.")
else:
    print("Yorum: Fiyat ve teslimat sÃ¼resi arasÄ±nda gÃ¼Ã§lÃ¼ bir iliÅŸki vardÄ±r.")

# 2. Interpret correlation between 'price' and 'review_score' - siparis fiyati ve memnuniyet
price_review_corr = correlation_matrix.loc['price', 'review_score']
print(f"\nKorelasyon (price vs. review_score): {price_review_corr:.2f}")
if abs(price_review_corr) < 0.1:
    print("Yorum: Fiyat ve memnuniyet puanÄ± arasÄ±nda Ã§ok zayÄ±f bir iliÅŸki vardÄ±r.")
elif abs(price_review_corr) < 0.4:
    print("Yorum: Fiyat ve memnuniyet puanÄ± arasÄ±nda zayÄ±f bir iliÅŸki vardÄ±r.")
elif abs(price_review_corr) < 0.7:
    print("Yorum: Fiyat ve memnuniyet puanÄ± arasÄ±nda orta dÃ¼zeyde bir iliÅŸki vardÄ±r.")
else:
    print("Yorum: Fiyat ve memnuniyet puanÄ± arasÄ±nda gÃ¼Ã§lÃ¼ bir iliÅŸki vardÄ±r.")

# 3. Interpret correlation between 'shipping_time' and 'review_score' - kargo suresi ve memnuniyet
shipping_review_corr = correlation_matrix.loc['shipping_time', 'review_score']
print(f"\nKorelasyon (shipping_time vs. review_score): {shipping_review_corr:.2f}")
if abs(shipping_review_corr) < 0.1:
    print("Yorum: Kargo sÃ¼resi ve memnuniyet puanÄ± arasÄ±nda Ã§ok zayÄ±f bir iliÅŸki vardÄ±r.")
elif abs(shipping_review_corr) < 0.4:
    print("Yorum: Kargo sÃ¼resi ve memnuniyet puanÄ± arasÄ±nda zayÄ±f bir iliÅŸki vardÄ±r.")
elif abs(shipping_review_corr) < 0.7:
    print("Yorum: Kargo sÃ¼resi ve memnuniyet puanÄ± arasÄ±nda orta dÃ¼zeyde bir iliÅŸki vardÄ±r.")
else:
    print("Yorum: Kargo sÃ¼resi ve memnuniyet puanÄ± arasÄ±nda gÃ¼Ã§lÃ¼ bir iliÅŸki vardÄ±r.")
```

**Hipotez (A/B) Testleri**

**Teslimat SÃ¼resinin Memnuniyete Etkisi**

AmaÃ§:
SipariÅŸin mÃ¼ÅŸteriye ulaÅŸtÄ±ÄŸÄ± sÃ¼re arttÄ±kÃ§a memnuniyet puanÄ± dÃ¼ÅŸÃ¼yor mu?

Hipotez:

H0: Teslimat sÃ¼resi memnuniyeti etkilemez

H1: Daha hÄ±zlÄ± teslimat daha yÃ¼ksek memnuniyet saÄŸlar

```python
# Review tablosunu yÃ¼kle
reviews = pd.read_csv("/content/olist_order_reviews_dataset.csv")

# Tarihleri datetime formatÄ±na Ã§evir
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

# Teslimat sÃ¼resi hesapla (gÃ¼n)
orders['delivery_days'] = (
    orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
).dt.days

# BirleÅŸtirme: Review puanlarÄ± ile teslimat sÃ¼resi
df_test1 = orders.merge(
    reviews[['order_id', 'review_score']],
    on="order_id",
    how="inner"
)

# Eksik veya negatif deÄŸerleri Ã§Ä±kar
df_test1 = df_test1[(df_test1['delivery_days'] >= 0) & (~df_test1['review_score'].isna())]
df_test1.head()
```

```python
fast_delivery = df_test1[df_test1["review_score"] >= 4]["delivery_days"]
slow_delivery = df_test1[df_test1["review_score"] <= 3]["delivery_days"]
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(fast_delivery, slow_delivery, equal_var=False)

print(f"T istatistiÄŸi: {round(t_stat, 2)}")
print(f"p deÄŸeri: {p_value}")
```

Ä°ki baÄŸÄ±msÄ±z grup arasÄ±nda anlamlÄ± bir fark olup olmadÄ±ÄŸÄ±nÄ± test etmek istediÄŸimiz iÃ§in t-testini kullandÄ±k. YapÄ±lan t-testi sonucunda p-deÄŸeri 0.05â€™ten Ã§ok daha kÃ¼Ã§Ã¼k bulunmuÅŸtur (p â‰ˆ 0).
Bu sonuÃ§, teslimat sÃ¼resinin mÃ¼ÅŸteri memnuniyeti Ã¼zerinde anlamlÄ± bir etkisi olduÄŸunu gÃ¶stermektedir.
Yani daha hÄ±zlÄ± teslim edilen sipariÅŸler daha yÃ¼ksek memnuniyet puanlarÄ± almaktadÄ±r.

Hipotezler:

H0: Kategoriler arasÄ±nda sipariÅŸ tutarlarÄ±nda anlamlÄ± fark yoktur

H1: En az bir kategori, sipariÅŸ tutarÄ± aÃ§Ä±sÄ±ndan anlamlÄ± ÅŸekilde farklÄ±dÄ±r

```python
from scipy.stats import f_oneway

# order_totals ve items_products
order_category = items_products.merge(order_totals, on='order_id', how='left')

# En popÃ¼ler ilk 5 kategori seÃ§imi
top_categories = (
    order_category['product_category_name']
    .value_counts()
    .head(5)
    .index
)

filtered_data = (
    order_category[order_category['product_category_name'].isin(top_categories)]
)

# GruplarÄ± oluÅŸtur
groups = [
    filtered_data[filtered_data['product_category_name'] == cat]['order_total_price']
    for cat in top_categories
]

# ANOVA testi
f_stat, p_value = f_oneway(*groups)

print(f"F istatistiÄŸi: {f_stat:.2f}")
print(f"p deÄŸeri: {p_value:.2e}")
```

**Ã–deme YÃ¶nteminin Tamamlanma OranÄ±na Etkisi**

AmaÃ§: SeÃ§ilen Ã¶deme yÃ¶ntemi, sipariÅŸin tamamlanÄ±p tamamlanmamasÄ±nÄ± etkiliyor mu?

Hipotez:

H0: Ã–deme yÃ¶ntemi tamamlanma oranÄ±nÄ± etkilemez.

H1: Ã–deme yÃ¶ntemi tamamlanma oranÄ±nÄ± etkiler.


```python
import pandas as pd
from scipy.stats import chi2_contingency

# Her sipariÅŸ iÃ§in en yÃ¼ksek payment_value iÃ§eren Ã¶deme yÃ¶ntemi temel alÄ±nÄ±r
dom_payment = (
    payments.sort_values(['order_id','payment_value'], ascending=[True, False])
            .drop_duplicates('order_id')[['order_id','payment_type']]
)

# SipariÅŸ durumundan tamamlandÄ± bilgisini Ã¼ret
orders_comp = orders[['order_id','order_status']].copy()
orders_comp['completed'] = (orders_comp['order_status'] == 'delivered').astype(int)

# BirleÅŸtir
pay_comp = orders_comp.merge(dom_payment, on='order_id', how='left')
pay_comp = pay_comp.dropna(subset=['payment_type'])

# Kontenjans tablosu: payment_type x completed
ct = pd.crosstab(pay_comp['payment_type'], pay_comp['completed'])

# Ki-kare testi
chi2, p, dof, expected = chi2_contingency(ct)

print(f"Ki-kare: {chi2:.2f}")
print(f"Serbestlik derecesi: {dof}")
print(f"p-deÄŸeri: {p:.2e}")
```

Ä°ki kategorik deÄŸiÅŸken arasÄ±ndaki anlamlÄ± iliÅŸkiyi incelediÄŸimiz iÃ§in Ki-kare testi kullandÄ±k. YapÄ±lan Ki-kare testi sonucunda p-deÄŸeri 0.05â€™ten Ã§ok daha kÃ¼Ã§Ã¼k bulunmuÅŸtur (p â‰ˆ 0).
Bu sonuÃ§, Ã¶deme yÃ¶nteminin sipariÅŸlerin tamamlanma oranÄ± Ã¼zerinde anlamlÄ± bir etkiye sahip olduÄŸunu gÃ¶stermektedir.

```python
completion_rate = (
    ct.div(ct.sum(axis=1), axis=0)[1]  # teslim edilenlerin oranÄ±
).reset_index()
completion_rate.columns = ['payment_type', 'completion_rate']

# not_defined'i Ã§Ä±kar
completion_rate = completion_rate[completion_rate['payment_type'] != 'not_defined']

fig = px.bar(
    completion_rate,
    x='payment_type',
    y='completion_rate',
    title='Ã–deme YÃ¶ntemine GÃ¶re Tamamlanma OranlarÄ±',
    labels={'completion_rate': 'Tamamlanma OranÄ±'}
)
fig.show()

```
